```python
import pandas as pd

data = {
    "Patient_ID": range(1, 51),
    "Age": [45, 52, 38, 60, 47, 55, 42, 50, 36, 58]*5,
    "Gender": ["Male", "Female"]*25,
    "Treatment_Group": ["Drug A", "Drug B"]*25,
    "Adverse_Event": ["Yes", "No", "No", "Yes", "No"]*10,
    "Outcome_Score": [78, 85, 90, 70, 88]*10
}

df = pd.DataFrame(data)
```

```python
df.groupby('Treatment_Group')['Outcome_Score'].mean()
```

```python
import matplotlib.pyplot as plt

df.groupby('Treatment_Group')['Outcome_Score'].mean().plot(kind='bar')
plt.xlabel("Treatment Group")
plt.ylabel("Mean Outcome Score")
plt.title("Treatment-wise Outcome Comparison")
plt.show()
```

```python
df['Adverse_Event'].value_counts().plot(kind='bar')
plt.xlabel("Adverse Event")
plt.ylabel("Number of Patients")
plt.title("Adverse Event Distribution")
plt.show()
```

Exploratory analysis showed balanced patient distribution across treatment groups. Drug B demonstrated a slightly higher mean outcome score compared to Drug A. Adverse event frequency was comparable between groups, suggesting acceptable tolerability. These findings indicate a favorable benefit-risk profile for Drug B in this simulated dataset.
